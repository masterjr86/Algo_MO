```python
def grad_descent(X, y, W, n, alpha):
  """Функция возвращает значения весов и значениие ошибки и подбирает значение
  коэффиицента скорости обучения при фиксированном значении количества итераций
  метода градиентного спуска
  """

  def calc_mse(y, y_pred):
    err = np.mean((y - y_pred) ** 2)
    return err

  for i in range(20):
    y_pred = W @ X
    err = calc_mse(y, y_pred)
    for ii in range(W.shape[0]):
        W[ii] -= alpha * (1 / n * 2 * np.sum(X[ii] * (y_pred - y)))
  print(W, err)


grad_descent(X, y, W, n, alpha)
```
